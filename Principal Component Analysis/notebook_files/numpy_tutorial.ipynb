### Be careful while copying arrays

```python
#changes will be made in both of them
a = np.array([1,23,98])
b = a
b[2] = 100
print("b :",b)
print("a :",a)
```

```python
#changes will be made in b
a = np.array([1,23,98])
b = a.copy()
b[2] = 100
print("b :",b)
print("a :",a)
```

```python
X = np.array([[0., 1., 1.], 
              [1., 2., 1.]])
X
```

```python
#Finding mean of a matrix
def mean(X):
    N,D = X.shape
    mean = np.sum(X,axis = 0)/N
    return mean
```

```python
#direct method
np.mean(X,axis = 0)
```

```python
#custom function
mean(X)
```

```python
#alternating method using loops
def mean_naive(X):
    N,D = X.shape
    mean = np.zeros(D)
    for n in range(D):
        mean[n] = np.sum(X[:,n])/N     
    return mean
```

```python
mean_naive(X)
```

```python
#direct method
def cov(X):
    covariance_matrix = np.cov(X,bias = True,rowvar = False)
    return covariance_matrix
```

```python
cov(X)
```

```python
def cov_naive(X):
    N, D = X.shape
    covariance = np.zeros((N, N))
    temp = X - mean(X)
    covariance = (temp.T@ temp)/N
    return covariance
```

```python
cov_naive(X)
```
